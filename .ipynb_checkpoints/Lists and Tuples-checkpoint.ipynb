### Challenge - Square List

Generate a list of squares of all numbers between 1 to 100. 
Do it in one line of code.

Output:
[1, 4, 9, 16, 25, 36, … ,10000]

````python
res = []
for i in range(1,101):
    res.append(i ** 2)
````

````python
res = [i**2 for i in range(1, 101)]
for x in res:
    print(x,end = ' ')
````

### Using map function

````python
def square(x):
    return x**2
res2 = list(map(square, range(1, 101)))
print(res2)
````

## Challenge - Even List
Given a list of random numbers, filter out all even numbers from it.
Do it in one line of code.

Input:
[27, 24, 21, 46, 48, 6, 25, 38, 10, 40]

Output:
[24, 46, 48, 6, 38, 10, 40]


````python
a = [27, 24, 21, 46, 48, 6, 25, 38, 10, 40]

b = []
for x in a:
    if x % 2 == 0:
        b.append(x)
````

````python
b = [x for x in a if x % 2 == 0]
````

````python
b
````

# Tuples
- Tuples are used to store multiple items in a single variable.
- A tuple is a collection which is ordered and immutable(unchangeable).
- Tuples are written with round brackets.
- Tuple items are indexed, the first item has index [0], the second item has index [1] etc.
- Tuples can be created with tuple constructor - `tuple()`
- Tuples can also be created by simply writing multiple values with comma separation.
- Tuples are processed faster than lists in Python.

````python
# Read-only list
continents = ('Asia', 'Africa', 'North America', 'South America', 'Europe', 'Antarctica', 'Australia')
````

## Challenge - Check if sorted

Given a list of numbers, check if it sorted (in increasing order).

````python
a = (1, 4, 10, 12, 8, 20)
````

````python
ans = "Sorted"

for i in range(len(a) - 1):
    if a[i] > a[i+1]:
        ans = "Not sorted"
        break
        
print(ans)
````

### Challenge - Write a function which returns the product of all its arguments.

The user should be able to call the function like

```
product(2, 3, 9, 1)    # 54
product(4, 2)          # 8
```

The function should be able to accept any number of arguments

````python
def product(*nums):
    print(nums, type(nums))
````

````python
product(10, 20, 30, 4, 2, 8, 10, 12)
````

````python
def speak(*words):
    print(words, type(words))
````

````python
speak("Hello", 'my', 'name', 'is', 'Tarun')
````

````python
def product(*args):
    p = 1
    print(args)
    for number in args:
        p = p * number
    return p
````

````python
product(10, 20, 7)
````

````python
product()
````

````python
a = [2,5,6,1,8]
````

````python
print(*a)

# * - spread operator
````
